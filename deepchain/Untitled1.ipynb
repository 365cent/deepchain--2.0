```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal(x,mu,gamma):
    y = np.exp(-(x-mu)**2/(2*gamma**2))/(gamma * np.sqrt(2*np.pi))
    return y
x  = np.linspace(-3,11,100)
y = normal(x,0,1)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
def sample_data(size,length = 100):
    data = []
    for _ in range(size):
        data.append(sorted(np.random.normal(4,1.5,length)))
    return np.array(data)

def random_sample(size,length = 100):
    data = []
    for _ in range(size):
        x = np.random.random(length)
        data.append(x)
    return np.array(data)
        
```

```python
plt.xlim(0,0.9)
plt.ylim(90,130)
x = np.linspace(0,1,100)
y = random_sample(1)
print(y)
plt.plot(x,random_sample(100,100))
plt.show()
```
